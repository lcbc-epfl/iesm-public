# Review of Linear Algebra Basics
To use quantum mechanics, you should be familiar with linear algebraoperations. The following exercises help you to review some basic operations with vector and matrices.

:::{admonition} Exercise 1
:class: exercise

 Calculate the vector product $\mathbf{c}=\mathbf{a}\times\mathbf{b}$
    with 

   $$
    \begin{aligned}
     \mathbf{a} = \left(\begin{matrix}
     2 \\ 6 \\ 4
     \end{matrix}\right)
     \quad
     \mathbf{b} = \left(\begin{matrix}
     5 \\ 1 \\ 7
     \end{matrix}\right)
     \end{aligned}
    $$

  and, for the same $\mathbf{a},\mathbf{b}$, the scalar product
    
   $$
    \begin{aligned}
     d = \mathbf{a}\cdot \mathbf{b}
     \end{aligned}
   $$
:::
